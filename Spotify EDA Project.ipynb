```python
# converting duration from milliseconds to seconds

df1['duration']= df1['duration_ms'].apply(lambda x: round(x/1000))
df1.drop('duration_ms', inplace=True, axis=1)
df1.head()
```

```python
df1.duration.head()
```


**Correlation Heatmap**

```python
corr_df= df1.drop(['id', 'name', 'explicit', 'artists', 'id_artists',
                   'release_date','key','mode'], axis=1).corr(method = 'pearson')

plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df, annot=True, fmt='.1g', vmin= -1, center= 0, vmax= 1, cmap='inferno', linewidths=1,linecolor='Black')
heatmap.set_title('Correlation heatmap between variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```


**Creating Regression Plots**

```python
sample_df= df1.sample(int(0.004*len(df1)))
print(len(sample_df))
```

```python
plt.figure(figsize=(10,7))
sns.regplot(data= sample_df, x= 'energy', y= 'loudness', color= 'c').set(title= 'Energy vs Loudness correlation')
```

```python
plt.figure(figsize=(10,7))
sns.regplot(data= sample_df, x= 'popularity', y= 'acousticness', color= 'r').set(title= 'Popularity vs acousticness correlation')
```
